## Exercise 1

We need to rank the following functions by their order of growth: $2^{\lg*(n)}, n^3, \lg*(n), e^n, n, n^2, \lg(n!), 2^n, n!, 2^{2^n}, \ln(n), (n + 1)!, n\lg(n), 1, 2^{2^{n + 1}}$.

Firstly, let's sort the functions with obvious classes of asymptotic complexity (information on that can be easily
found directly in lecture slides). Namely, constant < logarithmic < linear < n-log-n < quadratic < cubic <
exponential < factorial. We get:

$1 < \ln(n) < n < n\lg(n) < n^2 < n^3 < 2^n < n!$

Proving that $e^n$ grows faster than $2^n$ is quite easy (their ratio = const). Let's determine whether $(n + 1)!$ has the
higher order than $n!$.

$\lim_{n \to +\infty} \frac{n!}{(n + 1)!} = \lim_{n \to +\infty} \frac{n!}{n!(n + 1)} = \lim_{n \to +\infty} \frac{1}{n + 1} = 0$.

Finally, we get that $n! = o((n + 1)!)$.

$1 < \ln(n) < n < n\lg(n) < n^2 < n^3 < 2^n < e^n < n! < (n + 1)!$

$\log^*(n)$, or iterated logarithm, is a function that iteratively applies log to an argument until it gets below one (e.g.: log(log(log(...log(n)))), the result is the number of such applications. $\log^*(n)$ grows very slowly, much slower than $\log(n)$. The number of logs is far smaller that the original argument to which all of them were applied. We can confidently (though intuitively) say that $\log^*(n) = o(\log(n))$. Moreover, we conclude that $2^{\lg^*(n)} = o(\log(n))$. At last, it's obvious that $\lg^*(n) = o(2^{\lg^*(n)})$, as in case of a simple logarithm (both these functions increase monotonically). In the end, we modify the ranking:

$1 < \lg^*(n) < 2^{\lg^*(n)} < \ln(n) < n < n\lg(n) < n^2 < n^3 < 2^n < e^n < n! < (n + 1)!$

Using Stirling's formula, we have $\lg(n!) = O(n\lg(n))$. Change the ranking respectively:

$1 < \lg^*(n) < 2^{\lg^*(n)} < \ln(n) < n < n\lg(n) = \lg(n!) < n^2 < n^3 < 2^n < e^n < n! < (n + 1)!$

There are two functions left: $2^{2^n}$ and $2^{2^{n + 1}}$. It's quite easy to prove that $2^{2^n} = o(2^{2^{n + 1}})$, so let's save some space for proving that $(n + 1)! = o(2^{2^n})$.

$\lim_{n \to +\infty} \frac{(n + 1)!}{2^{2^n}} = \lim_{n \to +\infty} \frac{2^{\log_2((n + 1)!)}}{2^{2^n}} = \lim_{n \to +\infty} \frac{2^{(n + 1)\log_2(n + 1)}}{2^{2^n}} = \lim_{n \to +\infty} 2^{(n + 1)\log_2(n + 1) - 2^n} = |n\log_2(n) = o(2^n)| = 0$.

Finally, we get the answer:

$1 < \lg^*(n) < 2^{\lg^*(n)} < \ln(n) < n < n\lg(n) = \lg(n!) < n^2 < n^3 < 2^n < e^n < n! < (n + 1)! < 2^{2^n} < 2^{2^{n + 1}}$

Or in reverse (as the task requires):

$2^{2^{n + 1}} > 2^{2^n} > (n + 1)! > n! > e^n > 2^n > n^3 > n^2 > \lg(n!) = n\lg(n) > n > \ln(n) > 2^{\lg^*(n)} > \lg^*(n) > 1$

## Exercise 2

Let's plot 9 functions from the previous exercise. We will use log-scale in order to avoid some of them flatten out.

```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
def log_iterated(arg, base):
    n = 0
    
    while arg >= 1:
        n += 1
        arg = math.log(arg, base)
 
    return n
```

```python
n = range(5, 50)

plt.rcParams["figure.figsize"] = (10,8)

plt.plot(n, [log_iterated(x, 10) for x in n], 'blue', label='lg*(n)')                         # lg*(n)
plt.plot(n, [2 ** log_iterated(x, 10) for x in n], 'red', linestyle='--', label='2^lg*(n)')   # 2^lg*(n)
plt.plot(n, [math.log(x) for x in n], 'green', linestyle='-.', label='ln(n)')                 # ln(n)
plt.plot(n, [x * math.log(x, 10) for x in n], 'orange', linestyle=':', label='n lg(n)')       # n lg(n)
plt.plot(n, [math.log(math.factorial(x), 10) for x in n], 'purple', label='lg(n!)')           # lg(n!)
plt.plot(n, [x ** 2 for x in n], 'brown', linestyle='--', label='n^2')                        # n^2
plt.plot(n, [x ** 3 for x in n], 'olive', linestyle='-.', label='n^3')                        # n^3
plt.plot(n, [2 ** x for x in n], 'fuchsia', linestyle=':', label='2^n')                       # 2^n
plt.plot(n, [math.exp(x) for x in n], 'cyan', label='e^n')                                    # e^n

plt.yscale('log')
plt.xscale('log')

plt.legend()
plt.grid(True)
plt.xlabel('n')
plt.ylabel('f(n)')
plt.title('Function comparison (log-scale)')

plt.show()
```

We see that our growth ranking from the previous task is correct. $\ln(n)$ may not clearly outgrow $2^{\lg^*(n)}$ for such small $n$, but with $n \to \infty$ it will.

The recurrent equation $A(N) = A(N/2) + 1$ may correspont to the binary search algorithm.  
The recurrent equation $B(N) = 2B(N/2) + N$ may correspont to the merge sort.

In order to justify complexities, let's use the master method.

1. $A(N) = A(N/2) + 1$.  
$a = 1, b = 2 \implies N^{\log_21} = N^0 = 1; f(N) = 1.$  
Case 2: $f(N) = \Theta(\lg^0N) = \Theta(1)$, that is, $k = 0$.  
$A(N) = \Theta(\lg N) = O(\lg N)$.  
  
  
2. $B(N) = 2B(N/2) + N$.  
$a = 2, b = 2 \implies N^{\log_22} = N^1 = N; f(N) = N.$  
Case 2: $f(N) = \Theta(N\lg^0N) = \Theta(N)$, that is, $k = 0$.  
$B(N) = \Theta(N\lg N) = O(N \lg N)$.

Let's write a program to calculate $T(N)$ for arbitraty $N$ (where $T(N)$ is either $A(N)$ or $B(N)$).

```python
def A(N):
    if N == 1:
        return 1
    return A(N // 2) + 1


def B(N):
    if N == 1:
        return 1
    return 2 * B(N // 2) + N
```

```python
print(f"A(50) = {A(50)}, should be equal to 6.")
print(f"B(50) = {B(50)}, should be equal to 276.")
```

## Exercise 4

Let's parse file words.txt into a list of words and look up the first 10 elements.

```python
def parse_words():
    words_list = []
    f = open("words.txt", "r")
    
    words_list = f.read().splitlines()
    
    return words_list
```

```python
words = parse_words()
print(words[:10])
```

Now let's implement linear and binary search algorithms and check their correctness by seeking words either present or absent in the list of words.

```python
def linear_search(array, target):
    """
    Finds index of target in array via linear search
    
    Returns:
        if target is in array: index of target in array, number of comparisons done
        otherwise: -1, number of comparisons done
    """
    for i in range(len(array)):
        if array[i] == target:
            return i, i + 1
        
    return -1, len(array)


def binary_search(array, target):
    """
    Finds index of target in array via binary search
    
    Returns:
        if target is in array: index of target in array, number of comparisons done
        otherwise: -1, number of comparisons done
    """
    left = 0
    right = len(array) - 1
    # auxillary variable to keep the number of comparisons
    n = 0
    
    while left <= right:
        i = left + (right - left) // 2
        n += 1
        
        if target < array[i]:
            right = i - 1
        elif target > array[i]:
            left = i + 1
        else:
            return i, n
        
    # if the target is missing in the array, return -1
    return -1, n
```

```python
print(f"Searching word 'Acapulco' via linear search: {linear_search(words, 'Acapulco')[0]}")
print(f"Searching word 'Acapulco' via binary search: {binary_search(words, 'Acapulco')[0]}\n")

print(f"Searching word 'None' via linear search: {linear_search(words, 'None')[0]}")
print(f"Searching word 'None' via binary search: {binary_search(words, 'None')[0]}")
```

## Exercise 5

Let's pick 9 words from the words list along with some word missing in the list (here we have 'None') and measure working time of our search algorithms as well as the number of comparison operations done for each word.

```python
import random
from statistics import mean
from time import perf_counter as timer
```

```python
def select_random_words(n_words):
    return random.choices(words, k=n_words)
```

```python
def do_experiment(search_func, array, samples):
        results = []
        
        for sample in samples:
            start = timer()
            res = search_func(array, sample)
            end = timer()
            elapsed = end - start
            results.append((res[0], elapsed, res[1]))
            
        return results
```

```python
samples = ['Hagiographa', 'cruddy', 'stingier', 'readjustments', 'output', "Bayeux's", 'lathers', 'deafer', 'decrease']
# add some word that is missing in the words list
samples.append('None')

results_linear = do_experiment(linear_search, words, samples)
results_binary = do_experiment(binary_search, words, samples)

fig, (ax1, ax2) = plt.subplots(2, figsize=(10, 12))
fig.suptitle('Search algorithms comparison', y=1.02)

labels = [samples[i] + '\n' + str(results_linear[i][0]) for i in range(len(samples))]
# the label locations
x = np.arange(len(samples))
width = 0.5

rects1 = ax1.bar(x, [result[2] for result in results_linear], width, color='blue')
ax1.set_title('Linear search')
ax1.set_ylabel('Number of comparison operations')
ax1.set_xticks(x, labels)
ax1.bar_label(rects1, padding=3)

rects2 = ax2.bar(x, [result[2] for result in results_binary], width, color='orange')
ax2.set_title('Binary search')
ax2.set_ylabel('Number of comparison operations')
ax2.set_xlabel('Word and its index in the list')
ax2.set_xticks(x, labels)
ax2.bar_label(rects2, padding=3)

fig.tight_layout()

plt.show()
```

As far as the number of comparison operations for each word does not change with repeated code calls, we built the plot above without benchmarking. Here we see that binary search requires far less comparisons than linear one. Linear search works faster if the target word is closer to the beginning of the words list. Whereas binary search works relatively fast for each input. Word 'cruddy' was found faster than the other words because it happenned to be right in the centre of one of the earlier binary splits. Finding the missing word 'None' took both methods the maximal possible number of comparisons: 99171 for linear search and 17 for binary search ($\log_2 99171 + 1 = 17$).

Now let's compare average durations of the algorithms on different number of words and build plots.

```python
def benchmark(n_tests, n_words):
    linear_results = []
    binary_results = []
            
    for _ in range(n_tests):
        test_data = select_random_words(n_words - 1)
        # add some new word that is missing in list
        test_data.append('None')
        
        linear_result = do_experiment(linear_search, words, test_data)
        linear_result = sum([row[1] for row in linear_result]),\
                        sum([row[2] for row in linear_result])
        linear_results.append(linear_result)
        
        binary_result = do_experiment(binary_search, words, test_data)
        binary_result = sum([row[1] for row in binary_result]),\
                        sum([row[2] for row in binary_result])
        binary_results.append(binary_result)
        
    linear_results = mean([row[0] for row in linear_results]),\
                     mean([row[1] for row in linear_results])
    binary_results = mean([row[0] for row in binary_results]),\
                     mean([row[1] for row in binary_results])
        
    return linear_results, binary_results
```

```python
n_tests = 5
n_words = range(0, 101, 5)
linear_results = []
binary_results = []

for i in range(len(n_words)):
    linear_result, binary_result = benchmark(n_tests, n_words[i])
    linear_results.append(linear_result)
    binary_results.append(binary_result)
```

```python
fig, (ax1, ax2) = plt.subplots(2, sharex=True, figsize=(9, 10))
fig.suptitle('Time efficiency comparison of linear and binary search', y=0.95)

ax1.plot(n_words, [row[0] for row in linear_results], 'red')
ax1.set_title('Linear search')
ax1.set_ylabel('Average search duration (seconds)')
ax1.grid(True)

ax2.plot(n_words, [row[0] for row in binary_results], 'blue')
ax2.set_title('Binary search')
ax2.set_xlabel('Number of words being searched')
ax2.set_ylabel('Average search duration (seconds)')
ax2.grid(True)

plt.show()
```

From the plots, we see that binary search algorithm is much more faster than linear one. Approximately 500 times faster.

Let's devise an experiment to measure the cache influence on duration of binary search. We will use 'cachetools' module and TTLCache.

```python
import operator
from cachetools import cachedmethod, TTLCache
```

```python
class BinarySearch():
    
    def __init__(self, cachesize):
        # create cache
        self.cache = TTLCache(maxsize=cachesize, ttl=10e5)
        
    # the results of the method will be cached, the key value is target    
    @cachedmethod(operator.attrgetter('cache'))
    def execute(self, target):
        """
        Finds index of word in list via binary search

        Returns:
            if word is in list: index of word in list
            otherwise: -1
        """
        left = 0
        right = len(words) - 1

        while left <= right:
            i = left + (right - left) // 2

            if target < words[i]:
                right = i - 1
            elif target > words[i]:
                left = i + 1
            else:
                return i

        # if the target is missing in the array, return -1
        return -1
```

```python
def measure_search_time(search, n_tests, threshold):
    elapsed = 0
    
    for i in range(n_tests):
        # executing search of random elements with indices in [0, threshold]
        word = words[random.randint(0, threshold)]
        start = timer()
        search.execute(word)
        end = timer()
        elapsed += end - start
    
    # return mean of search time per word
    return elapsed / n_tests
```

Let's conduct a series of experiments:
1. Create cache of small size, execute searches on small area of the list, measure average time to find a word; repeat the searches on a large area. Results must differ because only elements of small area were cached.
2. Disable cache, repeat the experiments. Search times should be equal on small and large areas.
3. Create cache of large size, execute search on large area. Results should be better than before because more elements were cached.

```python
# create small cache of max size 1000
search = BinarySearch(1000)
print('Created cache of size 1000')
print(f"Current cash size: {search.cache.currsize}")

elapsed = measure_search_time(search, 10000, 1000)
print(f"Current cash size: {search.cache.currsize}")
print(f"Average duration of search on small cached area (n=1000): {elapsed} s")

elapsed = measure_search_time(search, 10000, 10000)
print(f"Average duration of search on large uncached area (n=10000): {elapsed} s")
```

We see that average duration of search was smaller on a cached area of the list. As soon as we started to seek elements on a larger uncached area, the results became almost three times worse.

```python
# disable cache
search = BinarySearch(0)
print('Disabled cache')
print(f"Current cash size: {search.cache.currsize}")

elapsed = measure_search_time(search, 10000, 1000)
print(f"Current cash size: {search.cache.currsize}")
print(f"Average duration of search on small area (n=1000): {elapsed}")

elapsed = measure_search_time(search, 10000, 10000)
print(f"Average duration of search on large area (n=10000): {elapsed}")
```

With cache disabled, average duration of search on both areas became almost equal. The searches take more time for the small area, they are almost two times slower. But search times for larger area improved, that's because there were no look-ups into the cache (which apparently take some time), it was empty.

```python
# create large cache of max size 10000
search = BinarySearch(10000)
print('Created cache of size 10000')

elapsed = measure_search_time(search, 100000, 10000)
print(f"Current cash size: {search.cache.currsize}")
print(f"Average duration of search on large cached area (n=10000): {elapsed / 10}")
```

Finally, we executed searches on large area with large cache. All elements in the area became cached. Now the search is executed 20 times faster.

### Conclusions

Cache generally speeds up running time of algorithms (we got boost in speed in up to 20 times). Nevertheless, if the cached elements are inquired too rarely, the running time decreases due to the useless calls to cache.
