```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
# in order to reduce the number of computations, find the logarithm of Catalan function inplace
# by calculating the sum of logarithms of base 2
def catalan_number_log(n):
    sum = 0
    
    for k in range(2, n + 1):
        sum += math.log2((n + k) / k)
    
    return sum
```

```python
xs = range(1, 1000, 5)

plt.rcParams["figure.figsize"] = (12, 10)

plt.plot(xs, [catalan_number_log(i) for i in xs], label='Catalon numbers')
# log_2(2^(2n)) = 2n
plt.plot(xs, [2 * i for i in xs], label='2^(2n)')

plt.title('Number of Different Trees of Size n')
plt.xlabel('n (number of nodes)')
plt.ylabel('value (log scale)')
plt.legend()

plt.show()
```

From the plot, it is obvious that both functions have the same growth rate, i.e. are in $\Theta$ relationship.

**Post-Order Case**  
We can see a pattern in post-order representation of the binomial tree. Let's illustrate it on the reference node 1011:
* $child(1011, i) = 1011 \oplus (0010 \gg i)$
* $parent(1011) = 1011 \oplus 0100$
  
So, for an arbitraty node, the index of its i-th child can be found by setting the i-th '1' from the end to '0'; the index of the parent can be found by setting the first zero from the beginning to '1'.
  
These ideas are implemented below (the code can be made into formulas, but they are too long to display).

```python
# returns 0000 if the given radix in n is 0, and 1111 if the given radix in n is 1
def mask(n, radix):
    flag = n & (2 ** radix)
    flag >>= radix
    for _ in range(3):
        flag |= flag << 1
    return flag

# finds the i-th '1' from the end and sets it to '0'
def postorder_child(n, i):
    return n & ~(((~n & 1) | (mask(n, 0) & ((~n & 2) | (mask(n, 1) & ((~n & 4) | (mask(n, 2) & (~n & 8))))))) >> (i + 1))

# finds the first '0' from the beginning and sets it to '1'
def postorder_parent(n):
    return ((~n & 1) | n) | (mask(n, 0) & (((~n & 2) | n) | (mask(n, 1) & (((~n & 4) | n) | (mask(n, 2) & ((~n & 8) | n))))))
```

```python
print("1st child of 0111 is: {0:b}.".format(postorder_child(int('0111', 2), 0)))
print("3rd child of 0111 is: {0:b}.".format(postorder_child(int('0111', 2), 2)))

print("\n1st child of 1011 is: {0:b}.".format(postorder_child(int('1011', 2), 0)))
print("2nd child of 1011 is: {0:b}.".format(postorder_child(int('1011', 2), 1)))

print("\nParent of 1011 is: {0:b}.".format(postorder_parent(int('1011', 2))))
print("Parent of 0101 is: {0:b}.".format(postorder_parent(int('0101', 2))))
```

## Exercise 3

**Level-Order Degree Sequence (LOUDS)** can be obtained by the following steps:
* calculate the degree of each node;
* write the degrees down in level order, separating them with 0;
* convert all the degrees to unary.
  
Each '1' represents a node of the tree. Then the **parent** of k is the number of 0's up to the k-th 1. The **children** of k are stored after the k-th 0.

```python
tree = '101111011101101001101001000100000'
```

```python
# returns a node under the k-th index (calculates the number of ones up to k-th entry)
def rank(tree, k):
    counter = 0
    
    for i in range(k):
        if tree[i] == '1':
            counter += 1
    
    return counter

# return an index of the node
def select(tree, node):
    counter = 0
    i = 0
    
    while counter != node:
        if tree[i] == '1':
            counter += 1
        i += 1
    
    return i


def parent(tree, node):
    return select(tree, node) - node


def child(tree, node, i):
    counter = 0
    k = 0
    
    while counter != node:
        if tree[k] == '0':
            counter += 1
        k += 1
    
    return rank(tree, k + i)
```

```python
print(f"Parent of 3 is {parent(tree, 3)}.")
print(f"Parent of 14 is {parent(tree, 14)}.")
print(f"2nd child of 1 is {child(tree, 1, 2)}.")
print(f"3rd child of 2 is {child(tree, 2, 3)}.")
```

**Parentheses Tree Representation**  
Parentheses Tree Representation can be obtained by traversing a tree in pre-order and writing down '(' each time we enter the node and ')' each time we leave the node. Then **parent** is an enclosing parentheses, the **first child** is a next parentheses (if exists), **next sibling** is an open parentheses following the matching closing parentheses (if exists).

<table>
    <tr>
        <th style="text-align:center">Pre-Order</th>
    </tr>
    <tr>
        <td><img src="Parentheses/1.png"></td>
     </tr>
</table>

```python
tree_str = '((((())())(())())((())())(())())'
```

```python
class ParenthesesTree:
    
    def __init__(self, string):
        self.tree = string

    
    def rank(self, pos):
        if pos >= len(self.tree) or self.tree[pos] != '(':
            return -1

        counter = -1

        for i in range(pos + 1):
            if self.tree[i] == '(':
                counter += 1

        return counter
    
    
    def select(self, node):
        counter = -1

        for i in range(len(self.tree)):
            if self.tree[i] == '(':
                counter += 1
            if counter == node:
                break

        return i
    
    
    def __find_enclosing_bracket(self, pos):
        flag = 0

        if pos == 0:
            return -1

        while flag != 1:
            pos -= 1
            if self.tree[pos] == '(':
                flag += 1
            else:
                flag -= 1

        return pos
    
    
    def parent(self, node):
        pos = self.select(node)
        return self.rank(self.__find_enclosing_bracket(pos))
    
    
    def first_child(self, node):
        return self.rank(self.select(node) + 1)
    
    
    def __find_closing_bracket(self, pos):
        flag = -1

        while pos < len(self.tree) and flag != 0:
            pos += 1
            if self.tree[pos] == '(':
                flag -= 1
            else:
                flag += 1

        return pos
    
    
    def next_sibling(self, node):
        pos = self.select(node)
        return self.rank(self.__find_closing_bracket(pos) + 1)
    
    
    def subtree_size(self, node):
        pos = self.select(node)
        closing = self.__find_closing_bracket(pos)
        return (closing - pos - 1) // 2
    
    
    def lowest_common_ancestor(self, first, second):
        if first == 0 or second == 0:
            return 0
        
        fparent = self.__find_enclosing_bracket(self.select(first))
        sparent = self.__find_enclosing_bracket(self.select(second))
        
        while fparent != sparent:
            if fparent < sparent:
                sparent = self.__find_enclosing_bracket(sparent)
            else:
                fparent = self.__find_enclosing_bracket(fparent)
        
        return self.rank(fparent)
```

```python
tree = ParenthesesTree(tree_str)
```

```python
print(tree.rank(tree.select(11)))
```

```python
print(tree.parent(11))
print(tree.parent(0))
```

```python
print(tree.first_child(9))
print(tree.first_child(8))
```

```python
print(tree.next_sibling(1))
print(tree.next_sibling(8))
```

## Exercise 5

In parentheses tree representation, the **subtree size** for some node is a half of the number of parentheses between the pair of parentheses, representing the node. The **lowest common ancestor** can be found by comparing the enclosing parentheses of both nodes (and then of parents of these nodes) until we come (or not come) to the same one.

```python
print(tree.subtree_size(0))
print(tree.subtree_size(9))
print(tree.subtree_size(11))
```

```python
print(tree.lowest_common_ancestor(4, 15))
print(tree.lowest_common_ancestor(5, 8))
```

Succinct data structures helps us save space when it comes to big amounts of data, e.g. DNA sequences, or small devices with limited memory. We can therefore achieve the data compression but the queries may not be supported efficiently. Besides, some operations may not be supported. So we succeed in memory (by definition, $(Z + o(Z))$ bits of space) but may lose in time efficiency and/or lose the possibility some operations.
